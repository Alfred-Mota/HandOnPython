Limpando o conjunto de dados

```python
df.drop(['id','Unnamed: 32' ], axis=1, inplace=True)
```

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix
```

```python
df.diagnosis = [1 if value == "M" else 0 for value in df.diagnosis]
df.head()
```

Separando a caracteristica alvo

```python
y = df.diagnosis

X = df.drop(['diagnosis'], axis=1, inplace=False)
```

```python
y_M = y == 1

print(f'Quantidade de celulas malignas {y_M.sum()}')
print(f'Porcentagem de celulas malignas {(y_M.sum()/len(y)).round(2)*100}%')
print(f'Quantidade de celulas benignas {len(y) - y_M.sum()}')
print(f'Porcentagem de celulas benignas {((len(y) - y_M.sum())/len(y)).round(2)*100}%')
```

Os dados possuem unidades muito diferentes, alguns na casa de milimetros e outros em cm. O modelo pode aprender que unidades maiores sao mais importantes, porem nem sempre é verdade e por isso necessitamos de uma Normalização

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
X_scaled = std_scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=.2, random_state=42)
```

```python
y_train_M = y_train == 1

print(f'Quantidade de celulas malignas no conjunto treino {y_train_M.sum()}')
print(f'Porcentagem de celulas malignas conjunto treino {(y_train_M.sum()/len(y_train)).round(2)*100}%')
print(f'Quantidade de celulas benignas conjunto treino {len(y_train) - y_train_M.sum()}')
print(f'Porcentagem de celulas benignas conjunto treino {((len(y_train) - y_train_M.sum())/len(y_train)).round(2)*100}%')
```

```python
from sklearn.linear_model import LogisticRegression

reg_log = LogisticRegression(random_state=42)
reg_log.fit(X_train, y_train)
```

```python
y_pred = reg_log.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score

print(confusion_matrix(y_test, y_pred))
print(precision_score(y_test, y_pred))#Referente aos Verdadeiros Positivos e Falsos Positivos
print(recall_score(y_test, y_pred))#Referente aos Verdadeiros Positivos e Falsos Negativos
print(accuracy_score(y_test, y_pred))#Referente aos Verdadeiros Positivos e Verdadeiros Negativos
```

Curva de Aprendizado

```python
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


def learning_curve(X,y,model):

    std_scaler = StandardScaler()
    X_scaled = std_scaler.fit_transform(X)

    X_train, X_val, y_train, y_val = train_test_split(X_scaled,y,test_size=.2, random_state=42)

    train_errors = []
    val_errors = []

    for m in range(2,len(X)):
        model.fit(X_train[:m], y_train[:m])
        y_pred_train = model.predict(X_train[:m])
        y_val_pred = model.predict(X_val)

        train_errors.append(root_mean_squared_error(y_pred_train, y_train[:m]))
        val_errors.append(root_mean_squared_error(y_val_pred, y_val))
    
    plt.plot((train_errors),'r+-', label='Train Errors', linewidth=2)
    plt.plot((val_errors),'b', label='Val Errors', linewidth=3)
    plt.xlabel("Test Size")
    plt.ylabel("RMSE")
    plt.grid()
    plt.legend()

```

```python
from sklearn.linear_model import LogisticRegression

reg_log = LogisticRegression(random_state=42)

learning_curve(X,y, reg_log)
```

```python
from sklearn.linear_model import LogisticRegression

#O INVERSO DE C É QUEM CONTROLA A REGULARIZAÇÃO
reg_log = LogisticRegression(random_state=42, penalty='elasticnet',solver='saga', l1_ratio=1, C=0.2)

learning_curve(X,y, reg_log)
```
